```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
dados = pd.read_csv('D:\Machine_Learning_HB\CSV\gaf_esp.csv', sep=';') # Estamos carregando o arquivo csv usando o ; como separador
```

```python
x = dados.drop(columns=['Espécie'])
x
```

```python
plt.scatter(x=dados['Comprimento do Abdômen'], y=dados['Comprimento das Antenas'])
```

```python
# Algoritmos de Agrupamento
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.mixture import GaussianMixture

# Avaliacao de desemepnho
from sklearn.metrics import adjusted_rand_score, silhouette_score
```

```python
# Modelos de agrupamento
rs=42
km = KMeans(n_clusters=4, random_state=rs)            # KNN k=4
gm = GaussianMixture(n_components=4, random_state=rs) # Gaussian Mixture k=4
db = DBSCAN(eps=0.4)
```

```python
# aplica o algoritmo e armazena o cluster de cada dado
km_c = km.fit_predict(x)  #
gm_c = gm.fit_predict(x)  # Permite mistura entre grupos (bem parecido com k-means)
db_c = db.fit_predict(x)  # Permite formatos não globulares
```

```python
km_c
```

```python
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import colors
import seaborn as sns
import scipy.cluster.hierarchy as sch   
```

```python
fig = plt.figure(figsize=(12, 8), dpi=80)
plt.subplot(2,2,1)
plt.scatter(x=dados['Comprimento do Abdômen'], y=dados['Comprimento das Antenas'])
plt.title('Original')
plt.subplot(2,2,2)
plt.scatter(x=dados['Comprimento do Abdômen'], y=dados['Comprimento das Antenas'], c=km_c)
plt.title('KMeans')
plt.subplot(2,2,3)
plt.scatter(x=dados['Comprimento do Abdômen'], y=dados['Comprimento das Antenas'], c=gm_c)
plt.title('GaussianMixture')
plt.subplot(2,2,4)
plt.scatter(x=dados['Comprimento do Abdômen'], y=dados['Comprimento das Antenas'], c=db_c)
plt.title('DBSCAN')
plt.subplots_adjust(hspace = 0.35)
```
