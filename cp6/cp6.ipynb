```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/CSV/heart_attack_prediction_dataset.csv", sep = ",")
```

```python
x = dados.drop(columns=["Patient ID","Continent","Hemisphere"])

x.head(10)
```

```python
import matplotlib.pyplot as plt
plt.hist(x['Age'], bins=20, edgecolor='k')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.title('Distribuição de Idade')
plt.show()
```

```python
sns.countplot(data=x, x='Sex')
plt.xlabel('Sexo')
plt.ylabel('Contagem')
plt.title('Contagem por Sexo')
plt.show()
```

```python
sns.boxplot(data=x, x='Sex', y='Cholesterol')
plt.xlabel('Sexo')
plt.ylabel('Colesterol')
plt.title('Distribuição de Colesterol por Sexo')
plt.show()
```

```python
x['Diabetes'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Proporção de Pacientes com Diabetes')
plt.show()
```

```python
x['Country'].value_counts().plot(kind='bar')
plt.xlabel('País')
plt.ylabel('Contagem')
plt.title('Distribuição de Pacientes por País')
plt.xticks(rotation=45)
plt.show()
```
