```python
import pandas as pd
```

```python
df = pd.read_csv('/content/household_power_consumption.txt', sep=';', low_memory=False)
```

```python
df.shape
```

```python
# Salva um novo dataframe somente com as colunas desejadas.
df1 = df.loc[:, ['Global_active_power', 'Sub_metering_2']]
df1.head()
```

```python
# Nota-se que são duas colunas que aparentam ser do tipo 'float', mas são 'object' (strings)
df1.info()
```

```python
# Se tentarmos converter em float64, encontraremos um erro...
df1 = df1.astype('float64')
df1
```

```python
# Vamos identificar os registros com o caractere "?" informado no erro acima....
df1[df1['Global_active_power'] == '?'].index
```

```python
# Em seguida, excluímos tais registros...
df1 = df1.drop(df[df['Global_active_power'] == '?'].index)
df1
```

```python
# Agora podemos converter em float64
df1 = df1.astype('float64')
df1
```

```python
# Agora, os registros estão no formato que desejamos :)
df1.info()
```
